```r
CorregirCampoMes  <- function( pcampo, pmeses )
{
  tbl <- dataset[  ,  list( "v_ant" = shift( get(pcampo), 1, type="lag" ),
                            "v_post" = shift( get(pcampo), 1, type="lead" ),
                            "v_foto_mes" = get("foto_mes"),
                            "v_orig" = get(pcampo)
                         ), 
                   by=numero_de_cliente ]
  
  tbl[ , promedio := ifelse(is.na(v_ant),v_post,ifelse(is.na(v_post),v_ant,(v_ant+v_post)/2)) ]
  tbl[ , nuevo_valor := ifelse(v_foto_mes %in% pmeses,promedio,v_orig)]
  
    keycol1 <-c("numero_de_cliente","v_foto_mes")
    setorderv(tbl, keycol1)  
  
      keycol2 <-c("numero_de_cliente","foto_mes")
    setorderv(dataset, keycol2)  
  
    
  dataset[ ,
           paste0(pcampo) := tbl$nuevo_valor ]
}
```
